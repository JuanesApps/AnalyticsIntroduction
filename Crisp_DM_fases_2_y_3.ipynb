# indexación, filtrado y asignación

**Pandas**

+ iloc: les permite hacer fitros a través de los indices númericos
+ loc: les permite hacer fitros a través de los indices como strings

Funciona como [fila, columna]

Ej: [0:10] : Va de 0 a 9, porque no se cuenta el último

```python
df.iloc[0:10,0:2]
```

```python
df.loc[df.Fare>=70]
```

```python
df.loc[df.Fare>=70].iloc[0:4,2]
```

# Cazando NaN

```python
df.info()
```

Dos variables con datos NaN
+ Cabin
+ Age

```python
df.loc[df.Age.isnull()]
```

```python
df.loc[df.Age.isnull(),'Age']
```

+ Asignar la mediana a las celdas de Age que aparecen como NaN

```python
df.loc[df.Age.isnull(),'Age'] = df.Age.mean()
```
